```python
%reset
from imutils import face_utils
import numpy as np
import imutils
import time
import dlib
import cv2
import datetime
   
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('C:/Users/ChungkiLee/Documents/GitHub/webcam/previewimgs/shape_predictor_68_face_landmarks.dat') # chage your file path

vs = cv2.VideoCapture(0)

(dt, micro) = datetime.datetime.now().strftime('%Y%m%d%H%M%S.%f').split('.')
b = float(dt) + float(micro)/1000000
time_x = []

# initialize hog + svm based face detector
hog_face_detector = dlib.get_frontal_face_detector()

# initialize cnn based face detector with the weights
# cnn_face_detector = dlib.cnn_face_detection_model_v1('C:/Users/ChungkiLee/mmod_human_face_detector.dat')

while True:
    ret, frame = vs.read()
    frame = imutils.resize(frame, width=640)
    frame = cv2.flip(frame, 1)
    gray  = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray  = cv2.pyrDown( gray )
    rects = detector(gray, 0)
        
    for rect in rects:
        shape0 = predictor(gray, rect)
        shape0 = face_utils.shape_to_np(shape0)
        
        mg = 3
        x, y = rect.left()*2, rect.top()*2
        w, h = (rect.right()*2 - x), (rect.bottom()*2 - y)
            
        cv2.rectangle(frame,(x-mg-2,y-mg-2),(x+w+mg+2,y+h+mg+2),(0,255,0),1)
    
        for (i, (x, y)) in enumerate(shape0):
            cv2.circle(frame, (x*2, y*2), 2, (0, 0, 255), 2)
 
    cv2.imshow("Frame", frame)
    (dt, micro) = datetime.datetime.now().strftime('%Y%m%d%H%M%S.%f').split('.')
    a = float(dt) + float(micro)/1000000
    time_x.append(a-b)
    if a-b >= 10:
        break
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break

# Closes all the frames
cv2.destroyAllWindows()
vs.release()
```
